```python
avg_rating=book_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating.rename(columns={'Book-Rating' : 'avg_ratings'}, inplace=True)
```

```python
popul_df = num_rating.merge(avg_rating, on='Book-Title')
```

# Collaborative Filtering bases recom_system

* Books should have atleast 50 ratings
* User should have rated atleast 200 books

```python
u=book_ratings.groupby('User-ID').count()['Book-Rating']>200
rating_users=u[u].index
```

```python
filtered_ratings=book_ratings[book_ratings['User-ID'].isin(rating_users)]
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
fam_books=y[y].index
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(fam_books)]
```

```python
p_table=final_ratings.pivot_table(index='Book-Title', columns='User-ID',values='Book-Rating').fillna(0)
```

```python
p_table
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_score=cosine_similarity(p_table)
```
